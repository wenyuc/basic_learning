```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = np.linspace(-5,5,100)
y1 = np.sin(x)
y2 = np.cos(x)

# create figure window
plt.figure(num = 3, figsize=(8,5))

# plot y1 and y2
plt.plot(x,y1)
plt.plot(x,y2, color = 'blue', linewidth = 5.0, linestyle = '--')

# set axes range
plt.xlim((-5,5))
plt.ylim((-2,2))

# set axes label
plt.xlabel("Data Sets", fontsize = 13, fontweight = 'bold')
plt.ylabel("Accuracy", fontsize = 13, fontweight = 'bold')

# set axes ticks 坐标轴刻度
my_x_ticks = np.arange(-5,5, 0.5)
my_y_ticks = np.arange(-2,2, 0.3)
plt.xticks(my_x_ticks, fontsize=8, color= 'red', rotation = 45)
plt.yticks(my_y_ticks, fontsize=8, color = 'red' )

# don't show axes ticks 不显示坐标刻度
# plt.xticks([])
# plt.yticks([])
# ax.set_xticks([])
# ax.set_yticks([])

# plt.axis('tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(1,13, 1)
y = range(1, 13, 1)

# plt.xticks()中第一个参数locs 表示x-axis刻度线显示标注的地方，即ticks放置的地方
# 第二个参数表示在locs数组表示的位置添加的标签，labels不赋值时，标签即为locs，如赋空值()则无ticks
plt.xticks(x)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import calendar

x = range(1,13,1)
y = range(1,13,1)

# ticks上的标签可以赋予任意其他的值
plt.xticks(x, calendar.month_name[1:13], color = 'blue',rotation = 60)
plt.plot(x,y)
plt.show()
```
