```python
df = nba.copy()    # df's modifications have no impact on nba.
df.shape
```

```python
df["difference"] = df.pts - df.opp_pts
df.shape
```

```python
df["difference"].max()
```

```python
# rename the verbose "game_result" and "game_location" columns 
df.rename(columns = {"game_result": "result", "game_location": "location"},
         inplace = True)
df.dtypes
```

```python
elo_columns = ["elo_i", "elo_n", "opp_elo_i", "opp_elo_n"]
df.drop(elo_columns, inplace= True, axis = 1)
df.shape
```

```python
df.info()
```

```python
df["date_game"] = pd.to_datetime(df["date_game"])
df.info()
```

```python
display(df["location"].nunique())
display(df["result"].nunique())
```

```python
df["location"].value_counts()
```

```python
# pandas provides the categorical data type for the same purpose
df["location"] = pd.Categorical(df["location"])
df.info()
```

```python
# Also convert the "result" column to categorical data type
df["result"] = pd.Categorical(df["result"])
df.shape
```

```python
df["notes"].fillna(
    value = "no notes at all", inplace = True)
df.info()
```
