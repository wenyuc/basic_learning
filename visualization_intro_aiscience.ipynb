```python
import matplotlib.pyplot as plt
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
import numpy as np
```

```python
x = np.linspace(0,10, 1000)
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
```

```python
x = np.linspace(0,10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 + y2
plt.plot(x,y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
plt.plot(x, y1, color = 'b', label = 'sin(x)')
plt.plot(x, y2, color = 'r', label = 'cos(x)')
plt.plot(x, y3, color = 'g', label = 'sin(x) + cos(x)')
plt.axis('tight')
plt.title('Sin and Cos plots')
plt.xlabel('X')
plt.ylabel('sin(x), cos(x), sin(x)+cos(x)')
plt.legend()
```

```python
# linestyle can be 'solid'(or '-'), 'dashed'(or '--'）, 'dashdot'(or '-.'), 'dotted'(or ':')
plt.plot(x, y, linestyle = '--')
```

```python
plt.plot(x, y1, '-.k')
```

```python
plt.plot(x, y1, '-.k')
plt.plot(x, y2, ':g')
plt.plot(x, y3, '--r')
plt.xlim(-2, 12)    # Add space in the x axis
plt.ylim(-2, 2)    # Add space in the y axis
```

```python
plt.plot(x, y1, '-.k')
plt.axis('tight')
```

```python
plt.plot(x, y1, '-.k')
plt.axis('equal')
```

```python
plt.plot(x, y1, '-.r')
plt.xlim(10, 0)
plt.ylim(1.5, -1.5)
```
