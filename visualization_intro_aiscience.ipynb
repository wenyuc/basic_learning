```python
x = np.linspace(0,10, 1000)
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
```

```python
x = np.linspace(0,10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 + y2
plt.plot(x,y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
plt.plot(x, y1, color = 'b', label = 'sin(x)')
plt.plot(x, y2, color = 'r', label = 'cos(x)')
plt.plot(x, y3, color = 'g', label = 'sin(x) + cos(x)')
plt.axis('tight')
plt.title('Sin and Cos plots')
plt.xlabel('X')
plt.ylabel('sin(x), cos(x), sin(x)+cos(x)')
plt.legend()
```

```python
# linestyle can be 'solid'(or '-'), 'dashed'(or '--', 'dashdot'(or '-.'), 'dotted'(or ':')
plt.plot(x, y, linestyle = '--')
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
```

```python
Z = 10*np.sin(X) + 2 * Y
plt.contourf(X, Y, Z, cmap = 'RdGy')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('contour plot')
```

# Density and pairplots from seaborn

```python
import seaborn as sns
plt.style.use('classic')
```

```python
X = 50 * np.random.randn(5000)
Y = 200 * np.random.randn(5000)
Z = 100 * np.random.randn(5000) + 500
plt.hist(X, density = True, alpha = 0.5)
plt.hist(Y, density = True, alpha = 0.5)
plt.hist(Z, density = True, alpha = 0.5)
```

```python
import pandas as pd

data = pd.DataFrame({'A':X, 'B':Y, 'C': Z})
```

```python
for col in data.columns:
    plt.hist(data[col], density = True, alpha = 0.5)
    sns.kdeplot(data[col], shade= True)
```

```python
for col in data.columns:
    sns.distplot(data[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot('A', 'B', data, kind ='kde')
```
