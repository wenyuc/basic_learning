### Scatter Plot 散点图

```python
import matplotlib.pyplot as plt
import numpy as np

N = 50

x = np.linspace(0., 10., N)
y = np.sin(x) ** 2 + np.cos(x)

plt.figure(figsize=(7,4.5))
# plt.scatter(x,y, s = 15, marker = 'x', label = r'$ y=sin^2(x)+cos(x) $', color = 'r')
plt.scatter(x,y, s = 15, marker = r'$\beta$', label = r'$ y=sin^2(x)+cos(x) $', color = 'r')
plt.axis('equal');    # x轴，y轴比例尺相同

plt.legend()
plt.xlabel(r'$x$ (rad)')
plt.ylabel(r'$y$')
plt.show()
```

```python
# 为x轴和y轴创建不同的函数
np.random.seed(100)
N = 50

randx = np.random.random(N) * 100
randy = np.random.random(N) * 100

plt.figure(figsize=(7,6))

plt.scatter(randx, randy, marker = 'o', s = 150, color = 'blue')
plt.axis('equal')
plt.xlabel('randx')
plt.ylabel('randy')

plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate data
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

#draw
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
