#### Example1 
To Operate or not to operate?
Proximal humerous fractures in the elderly

|Treatment  |Improved | Not Improved | Total |
|---|---|--- | ---|
|Operative|   62|    38|   100|
|Physio only | 48|  52|  100|
|Total|  110  |90|  200|

Conduct a full hypothesis test to answer the following questions:
(a) Is there evidence of a difference in outcomes?
(b) The Orthopaedic Society requires an outcome improvement of at least 
5% before updating treatment protocols. 
Should they recommend surgery for proximal humerous fractures?

Answer:

##### null hypothesis and level of significance
p1 = probability of positive outcome for operative group
p0 = probability of positive outcome for non-operative group

H0: theta = 0
Ha: theta <> 0 , let theta = p1-p0

If H0 is true, how will the sampling statistic be distributed?
Probability density function for sample difference, under null 
hypothesis (theta = 0). Sample difference (theta_hat)

E(theta_hat) = 0
V(thetha_hat) = V(p1-p0) = V(p1) + V(p0) = p1(1-p1)/n1 + p0(1-p0)/n0= p(1-p)/n1 + p(1-p)/n0  = p(1-p)(1/n1+1/n0)

**theta_hat ~ N(0, p(1-p)(1/n1 + 1/n0))**

alpha = level of significance = P(Type 1 Error)

##### Test Statistic #####
If theta_hat ~ N(0, p(1-p)(1/n1 + 1/n0))
consider the test statistics T = theta_hat / SE(theta_hat) = theta_hat / sqrt(p(1-p)(1/n1 + 1/n0))
T ~ N(0, 1)

critical value = 1.96
calculated test statistics t = theta_hat / sqrt(p_hat(1-p_hat)(1/n1 + 1/n0))

In the example p_hat = (62+48) / 200 = 0.55
n1=100, n0=100
theta_hat = 0.62-0.48 = 0.14
calculated t = 1.99 , right of 1,96, in the reject area, Reject H0!
Therefore, there is enough evidence to infer a difference between the two treatment options (at the 5% 
level of significance).

##### p-value
The **p-value** is the proportion of repeated samples, under the null hypothesis, that would be
as extreme as the test statistic we generated.
if p< alpha, then there is enough evidence to reject H0
if p> alpha, then there is NOT enough evidence to reject H0

As p=0.047 < 0.05, we can reject H0 at 5% level of significance. There is enough evidence to 
support a difference in proportions p1 and p0.

##### Confidence Interval
**Confidence intervals** are constructed around the sample statistics,theta_hat ( or 
calculated test statistic, t).

p1_hat = 0.62, p0_hat = 0.48 theta_hat = p1_hat - p0_hat = 0.14

se(theta_hat) = sqrt(p1(1-p1) / n1 + p0(1-p0)/n0)= 0.0697

95% CI: theta_hat +- se(theta_hat)*z0.975
        0.14 +- 0.0697(1.96) [0.0035, 0.2765]
=> We are 95% confidence that the interval from 0.35% to 27.65% contains the true
population difference theta.

##### Significant treatment difference
(b)The Orthopaedic Society requires an outcome improvement of at least 
5% before updating treatment protocols. 
We did :
H0: theta=0
H1: theta <>0

Actually, we need:
H0: theta less than or equal to 0.05
Ha: theta > 0.05



##### Power and sample size

|Do not reject H0 | Reject H0 |
|---|---|
|True Negative(1-alpha) | False positive, alpha, Type1 Error|     #H0 is true(no effect)
|False negative,beta, Type II error | True positive, 1-beta, Power|  #H0 is false(effect)

row 1: H0 is true(no effect); row 2: H0 is false (effect)

**The power of a model is the probability of rejecting a FALSE null hypothesis.**
**i.e., the ability of a model to detect a specified difference, delat. **
