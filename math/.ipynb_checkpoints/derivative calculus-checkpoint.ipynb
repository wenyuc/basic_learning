### triangle functions
$$ \cfrac {d}{d\mathit{x}} [sin(u(x))] = cos(u(x))u^{'}(x) \\
 \cfrac {d}{d\mathit{x}} [cos(u(x))] = -sin(u(x))u^{'}(x) \\
 \cfrac {d}{d\mathit{x}} [tan(u(x))] = sec^2(u(x))u^{'}(x) \\
 \cfrac {d}{d\mathit{x}} [cot(u(x))] = -csc^2(u(x))u^{'}(x) \\
 \cfrac {d}{d\mathit{x}} [sec(u(x))] = sec(u(x)) tan(u(x)) u^{'}(x) \\
 \cfrac {d}{d\mathit{x}} [csc(u(x))] = -csc(u(x)) cot(u(x)) u^{'}(x) $$

### Natural Log
$$ \cfrac{d}{dx}[ln(u(x))] = \cfrac{u^{'}(x)}{u(x)} $$
$$ \cfrac{d}{dx}[log_a^{u(x)}] = \cfrac{u^{'}(x)}{u(x) \cdot ln\;a} $$
For example:
$$ \cfrac{d}{dx}[ln(x^4-x^3)] = \cfrac{4x^3 - 3x^2}{x^4-x^3} $$

$$ \cfrac{d}{dx}[ln(sin(x))] = \cfrac{cos x}{sin x} = cot(x) $$

$$ \cfrac{d}{dx}[log_{5}^{(x^3+x^2)}] = \cfrac{3x^2 + 2x}{(x^3+x^2)\cdot ln5} $$

### Exponential function
$$ \cfrac{d}{dx}[e^{u(x)}]  = {e^{u(x)}} \cdot u^{'}(x) $$
$$ \cfrac{d}{dx}[a^{u(x)}]  = {a^{u(x)}} \cdot u^{'}(x) \cdot lna $$
For example:

$$ \cfrac{d}{dx}[e^{x^2}] = e^{x^2}\cdot (2x) \\
 \cfrac{d}{dx}[7^{x^2}] = 7^{x^2}\cdot (2x)\cdot ln7 $$

$$ y = x^x \\
 ln y = x\cdot lnx \\
 \cfrac{d}{dy}[ln y] \cdot \cfrac{dy}{dx} = \cfrac{d}{dx}[x\cdot ln x] \\
 \cfrac{1}{y} \cdot \cfrac{dy}{dx} = 1\cdot ln x + x \cdot \cfrac{1}{x} \\
 \cfrac{1}{y} \cdot \cfrac{dy}{dx} = 1\cdot ln x + 1 \\
 \cfrac{dy}{dx} = y \cdot (ln x + 1 )  \\
 \cfrac{dy}{dx} = x^x \cdot (ln x + 1 )  $$
 

$$ \cfrac{d}{dx}[u^v] = u^v\cdot[v^{'}ln u + v \cdot \cfrac{u^{'}}{u}] $$

$$ \cfrac{d}{dx}[x^{sin x}] = x^{sin x}\cdot[cos x \cdot ln x + sin x \cdot \cfrac{1}{x}]  $$

### Product function
$$ \cfrac {d}{dx} [u(x) \cdot v(x)] = u^{'}(x)\cdot v(x) + u(x) \cdot v^{'}(x) $$

### Quotient function
$$ \cfrac{d}{dx}\left[\cfrac{u(x)}{v(x)}\right] = \cfrac {v(x)\cdot u^{'}(x) - u(x)\cdot v^{'}(x)}{v(x)^2} $$

### Chain Rule
$\cfrac{d}{dx}[f(g(x))] = f^{'}(g(x))\cdot g^{'}(x) $

$\cfrac{d}{dx}[f(x)^n] = n\dot[f(x)]^{n-1} \dot f^{'}(x) $

### Implicit Differentiation

比如 $ y^2 + y = 3x^5 -7x $

如果能解出来，得到$ y = f(x) $，这就是显函数(explicit function).

如果不能解出来，无法得到$ y = f(x) $，这就是隐函数(implicit function)

无论能否解出$ y = f(x) $， y都是x的函数

所以，当$ y^2 + y = 3x^5 -7x $两边对x求导时，运用复合函数求导法(chain rule)，得到

$ 2y\left(\cfrac{dy}{dx}\right) + 1\cdot \cfrac{dy}{dx} = 15x^4 -7 $

$\cfrac{dy}{dx} = \cfrac{(15x^4 -7)}{(2y+1)} $


### Derivative of inverse function

$ y = x^3 +4x^2 +6 $
**solution 1**
$\left[f^{-1}(x)\right]^{'} = \cfrac {1}{f^{'}(y)} $
$ y^{'} = 3x^2+8x$

$\left[f^{-1}(x)\right]^{'} = \cfrac {1}{(3y^2+8y)} $

**solution 2**

swap x and y, becoming

$ x= y^3 +4y^2 +6 $

implicit derivative

$ 1 = 3y^2 \cfrac{dy}{dx} + 8y \cfrac{dy}{dx} $

$\left[f^{-1}(x)\right]^{'} = \cfrac {1}{(3y^2+8y)} $

### Limits

$ f^{'}(x) = \lim_{h\rightarrow 0} \cfrac {f(x+h) - f(x)}{h} $

#### L'hopital's Rule 
(https://www.youtube.com/watch?v=thEnl_gCjXA)

$ lim_{x\rightarrow c} \cfrac{f(x)}{g(x)} = lim_{x\rightarrow c} \cfrac{f^{'}(x)}{g^{'}(x)} $

For example 
1. 

$  lim_{x\rightarrow \infty} \cfrac{x^2}{e^x} =  lim_{x\rightarrow \infty} \cfrac{2x}{e^x} =  lim_{x\rightarrow \infty} \cfrac{2}{e^x} = 0$

2. 
$  lim_{x\rightarrow \infty} \cfrac{\ln x}{x} = lim_{x\rightarrow \infty} \cfrac{\frac{1}{x}}{1} = lim_{x\rightarrow \infty} \cfrac{1}{x} = 0 $

3.

$  lim_{x\rightarrow 0} \cfrac{sin(7x)}{sin(4x)} = lim_{x\rightarrow 0} \cfrac{cos(7x)\cdot 7}{cos(4x)\cdot 4} = lim_{x\rightarrow 0} \cfrac{cos(0)\cdot 7}{cos(0)\cdot 4} = \frac{7}{4} $

4.

$ y = lim_{x\rightarrow \infty} x^{\frac{1}{x}} $

$ \ln y = lim_{x\rightarrow \infty} \ln x^{\frac{1}{x}} $

$ \ln y = lim_{x\rightarrow \infty} \frac{1}{x} \ln x $

$ \ln y = lim_{x\rightarrow \infty} \cfrac{\ln x}{x} $

$ \ln y = lim_{x\rightarrow \infty} \cfrac{\frac{1}{x}}{1} = 0 $

$ y = e^0 = 1 $
