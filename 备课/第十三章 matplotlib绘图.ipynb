### 折线图

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

- 调整线条颜色 color /c = 

```python
offsets = np.linspace(0, np.pi, 5)
colors = ["blue", "g", "r", "yellow", "purple"]
for offset, color in zip(offsets, colors):
    plt.plot(x, np.sin(x-offset), color=color)    
```

- 调整线条风格 linestyle / ls=

```python
x = np.linspace(0,10, 11)
offset = list(range(8))
linestyles = ['solid', 'dashed', 'dashdot', 'dotted', '-', '--', '-.', ':']
print(len(linestyles))
for offset, linestyle in zip(offsets, linestyles):
    plt.plot(x, x+offset, linestyle=linestyle)
```

- 调整线条宽度 linewidth lw =

```python
x = np.linspace(0, 10, 11)
offset = list(range(0, 12,3))
linewidths = [i*2 for i in range(1, 5)]
for offset, linewidth in zip(offsets, linewidths):
    plt.plot(x, x+offset, linewidth = linewidth)
```

- 调整数据点标记 marker =

```python
x = np.linspace(0, 10, 11)
offsets = list(range(0, 12,3))
markers = ["*", "+","o", "s"]
for offset, marker in zip(offsets, markers):
    plt.plot(x, x+offset, marker=marker,markersize=10)
```

- 颜色和标记一起设置

```python
x = np.linspace(0,10,11)
offsets = list(range(0,8, 2))
color_linesytles = ['g--', 'b-', 'k--', 'r:']
for offset, color_linestyle in zip(offsets, color_linesytles):
    plt.plot(x, x+offset, color_linestyle)
```

加数据点

```python
x = np.linspace(0,10,11)
offsets = list(range(0,8, 2))
color_marker_linesytles = ['g*--', 'bo-', 'ks--', 'r+:']
for offset, color_marker_linestyle in zip(offsets, color_marker_linesytles):
    plt.plot(x, x+offset, color_marker_linestyle, markersize=20)
```

- 设置图形标签

```python
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x))
plt.title("A Sine Curve", fontsize=15)
plt.xlabel("x",fontsize=15)
plt.ylabel("sin(x)", fontsize=15)
```

```python
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x), "b-", label="Sin")
plt.plot(x, np.cos(x), "r--", label="Cos")
plt.ylim(-1.5, 2)
plt.legend(loc="best", frameon=True, fontsize=10)
```

- 添加文字

```python
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x), "b--")
plt.text(3, 0.5, "y=sin(x)", fontsize=10)
```

#### 添加箭头

```python
x = np.linspace(0, 2 * np.pi, 100)
plt.plot(x, np.sin(x), "b--")
plt.annotate("local minimum", xy=(1.5 * np.pi, -1), xytext=(4.5, 0), arrowprops=dict(facecolor="black", shrink=0.1))
```
