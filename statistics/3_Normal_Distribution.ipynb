### Gaussian distribution (Continuous) 
https://www.youtube.com/watch?v=CjF_yQ2N638

Gaussian distribution is a very common continuous probability distribution
$$ 
X \sim N(\mu, \sigma) \\
Z = \cfrac{X-\mu}{\sigma}\\
X = \mu + Z \cdot \sigma \\
$$
**P.D.F. of Gausse Distribution:**
$$
f(x) = \cfrac{1}{\sqrt{2\pi} \sigma} exp^{-\cfrac{1}{2}\left(\cfrac{x-\mu}{\sigma}\right)^2} \\
$$

**Empirical Rule**
- 68% of the x values lies within ONE *standard deviation* of the mean. 
- 95% of the x values lies within TWO *standard deviation* of the mean. 
- 99.7% the x values lies within THREE *standard deviation* of the mean.

#### Confidence Level and margin of error

1. The average test scores in a physics class is normally distributed with a standard deviation of 5.4. 50 scores with a sample mean of 79 were selected at random. 
(a) Find a 95% confidence interval for the population mean test score.

$ \sigma = 5.4, n=50, \bar x = 79$  How to get the population $\mu$ ?

$ \bar x - Z_\alpha\cfrac{\sigma}{\sqrt{n}} \leq \mu \leq \bar x + Z_\alpha\cfrac{\sigma}{\sqrt{n}} $

$ EBM = Z_\alpha\cfrac{\sigma}{\sqrt{n}} $

Therefore,

$ CI \rightarrow \left(\bar x - EBM, \bar x + EBM \right) $

$ A_L = \cfrac{CL + 1}{2} $, in which CL is Confidence Level (in our case, 0.95)

$ A_L = \cfrac{0.95+1}{2} = 0.975 $

Using approrpiate z_score table, $ Z_{\alpha} = 1.96 $

$ \sigma_{\bar x} = \cfrac{5.4}{\sqrt{50}} =0.76368 $

So, our 95% CI = $\rightarrow (79 \pm 1.96 \cdot 0.76368) $

(b) what is the value of margin error?
Margin Error is the $ EBM = Z_{\alpha}\cfrac{\sigma}{\sqrt{n}} = 1.4968 $

#### What is the difference between Confidence Interval and Confidence Level?

**The confidence Interval (CI)** is the range of values that we expect the **population mean** to be located in with certain probability.

When $CI \uparrow, EBM \uparrow$. When $ n \uparrow, EBM \downarrow $.
If want $EBM \downarrow, CI \downarrow or\; n  \uparrow $

**The Confidence Level (CL)** is the area under the curve in which we expect the **population mean** to be located in.

$ CL = 1- \alpha $, in which $\alpha$ is the probability that the mean is outside of the confidence interval. 



#### Student's T Distribution - Confidence Intervals & Margin of Error
1. The average weight of 20 students in a certain school was found to be 165 lbs with a standard deviation of 4.5. 
(a) construct a 95% confidence interval for population mean.

Population mean $ \mu \rightarrow \bar x \pm t_{\nu, \frac{\alpha}{2}} \cfrac{s}{\sqrt{n}} $

$ \nu = n-1 = 19 $, which represents the degrees of freedom. 

$ \cfrac{\alpha}{2} = \cfrac{1-CL}{2} = 0.025 $

$ \mathit t_{19, 0.025} =2.09302 $

The critical student_t distribution table: https://www.thoughtco.com/student-t-distribution-table-3126265

$ \mu \rightarrow (165 \pm 2.09302 \cdot \cfrac{4.5}{\sqrt{20}}) $

In Student t distribution, $EBM = t_{\nu, \frac{\alpha}{2}} \cfrac{s}{\sqrt{n}} $

95% $ CI \rightarrow (\bar x - EBM , \bar x + EBM) $

**In the calculation of standard deviation in student t distribution, use (n-1), not n.**


#### Finding the confidence interval of a population proportion Using the Normal Distribution

https://www.youtube.com/watch?v=_aQTgv_OOSU

1. 800 randomly selected college students at university XYZ were asked if they own a laptop. 584 responded yes to the survey.

**(a)** Can we approximate the binomial distribution with a normal distribution?

if $ n\cdot{p}\cdot(1-p) \geq 10 $, then the binomial distribution is a good approximation of the normal distribution. 

In the example, **we know $ n= 800, X = 584 $. But we dont know the $p$.** we dont know the exact proportion of all of the students at university XYZ if they own a laptop or not. However, **we can estimate the population proportion with a sample proportion**. 

$ p \rightarrow population\; proportion , \hat{p}\; or\;  p\prime \rightarrow sample\;proportion $

$ \hat p = \cfrac{x}{n} = \cfrac{584}{800} = 0.73 $

$ n\cdot{\hat p}\cdot{(1- \hat p)} = 800\cdot{0.73}\cdot{(1-0.73)} = 157.86 \geq 10 $

So, **this distribution can be approximated to a normal distribution**.

**(b)** Construct a 95% confidence interval to estimate the true population proportion of students at University XYZ who own laptops and determine the margin of error?

$ n = 500, x=584, \hat p = 0.73 $

Firstly, we need to calculate $\hat q$. $\hat p$ respesents the proportion for the number of success. $\hat q$ represents the proportion for the number of failures. $ \hat q = 1 - \hat p = 0.27 $

Secondly, we need to determine **the margin of error or the error bound for the population** once we have that we can construct the confidence interval.

$ EBP = Z \cdot \sqrt{\cfrac{\hat{p}\hat{q}}{n}} $

We can find the Z from the 95% confidence level. 

$ A_L = \cfrac{CL + 1}{2} = \cfrac{1+0.95}{2}= 0.975$

From z_scores table, we can find $ Z=1.96$.

$ EBP = 1.96 \cdot \sqrt{\cfrac{0.73\cdot 0.27}{800}} = 0.0308$

$ CI \rightarrow (\hat p - EBP, \hat p + EBP) $

$ CI \rightarrow (0.73 - 0.0308, 0.73 + 0.0308) $

**(c)** What sample size is needed to estimate true proportion with a 2% margin of error at a 90% confidence level?

$ EBP = 0.02, CL=0.90, Z=1.645 $

$ n = \cfrac{Z^2\cdot\hat{p}\cdot{\hat q}}{EBP^2} = 1333.39 \rightarrow  1334$

### Hypothesis Testing - Difference of Two Means - Student's Distribution & Normal Distribution

https://www.youtube.com/watch?v=UcZwyzwWU7o

1. A test was conducted on two different classes to see if there was any significant difference between the two teachers. 

The final scores of 15 students were sampled in the 1st class yielding a mean score of 82 with a standard deviation of 2.4

The final scores of 12 students were sampled in the 2nd class yielding a mean score of 84 with a standard deviation of 1.7.

Determine if there is any major difference at a 5% significance level. 

1st: $ n_1=15, \bar x_1=82, s_1=2.4$

2nd: $ n_2=12, \bar x_2=84, s_1=1.7$

$\alpha = 0.05$

$H_0: u_1=u_2$

$H_a: u_1 \neq u_2$

Use student distriubtion because $n<30$

$ df(degree of freedom) = \cfrac{\left(\cfrac{s_1^2}{n_1} + \cfrac{s_2^2}{n_2}\right)^2}{\cfrac{\left(\cfrac{s_1^2}{n_1}\right)^2}{n_1-1} + \cfrac{\left(\cfrac{s_2^2}{n_2}\right)^2}{n_2 - 1}} $

In the example, $ df= 24.7\approx 25$

$ t_{25, 0.025} = 2.059$

So what we need to do is to get our **calculated t value** and compare it to the **critical t value**. We want to see if our calculated t value lies in the rejection region or in the middle.

**calculated t value**

$ t_c = \cfrac{(\bar x_1 - \bar x_2) - (u_1 - u_2)}{\sqrt{\cfrac{s_1^2}{n_1} + \cfrac{s_2^2}{n_2}}} $

In the example, $t_c \approx -2.53$

$t_c$ in the rejection region because it less than -2.059

So we **reject** the null hypothesis.

2. A test showed that the old factory refines oil at a mean rate of 3.1L per second at a standard deviation of 1.0 using a sample of 40. 

The new factory refines oil at a mean rate of 3.8L per second at a standard deviation of 1.5 using a sample size of 36. 

Determine if there is any major difference at a 10% significance level.

old: $ n_1=40, \bar x_1=3.1, s_1 = 1$

new: $ n_2=36, \bar x_2 = 3.8, s_2 = 1.5$

$ \alpha = 0.10

$H_0: u_1=u_2$

$H_a: u_1 \neq u_2$

Use **Normal distribution** because sample size lager than 30.

$ A_L = 0.90 + \frac{1-0.90}{2} = 0.95 $

From z_scores table, $ Z = 1.645$

$ Z_c = \cfrac{(\bar x_1 - \bar x_2) - (u_1 - u_2)}{\sqrt{\cfrac{s_1^2}{n_1} + \cfrac{s_2^2}{n_2}}} $

In the example, $Z_c = -2.366$

$Z_c$ in the rejection region because it less than -1.645

So we **reject** the null hypothesis. The new factory is better than the old factory.

#### Hypothesis Testing - Solving Problems with Proportions

1.
A tech company believes that the percentage of residents in town XYZ that own a cell phone is 70%. A marketing manager believes this value to be different. He conducts a survey of 200 individuals and found that 130 responded yes to owning a cell phone.

**(a)** State the null and alternative hypothesis.

$ H_0 : p \; (proportion) = 0.70 $

$ H_a : p \neq 0.70 $

**(b)** At a 95% confidence level, is there enough evidence to reject the null hypothesis?

We know $ n=200, x=130$, then $\hat p = \frac{x}{n}=0.65$

Sothe proportion that's associated with the null hypothesis $ p_0 = 0.70 \; q_0 = 1-p_0 =0.30$

We have a two-tailed test.

We need the **calculated Z value** and the **critical Z value**.

$A_L = 0.95 + \frac{1-0.95}{2} = 0.975$

After checking the z_scores table, $Z_{\mathit{critical}} = 1.96$

So, the **critical values** is (-1.96, +1.96).

$ Z_{\mathit{calculated}} = \cfrac{\hat p - p_0}{\sqrt{\cfrac{p_0q_0}{n}}} $

In the example, $Z_{\mathit{calculated}} \approx -1.54 $

The $Z_{\mathit{calculated}}$ is in the critical region.
So, we **cannot reject** the null hypothesis. We dont have enough evidence to reject hypothesis.

So, we **reject** the null hypothesis. We have enough evidence to reject hypothesis.

#### Hypothesis Testing with Two Proportions

1.
In the first group, 32 out of 800 were found to contain some sort of defect. 

In the second group, 30 out of 500 were found to have a defect.

$ n_1=800, x_1=32, \hat p_1 = \frac{32}{800} = 0.04 $

$ n_2=500, x_2=30, \hat p_2 = \frac{30}{500} = 0.06 $

$ \alpha = 0.05, \cfrac{\alpha}{2} = 0.025$

$ H_0: p_1 = p_2 $ $H_a: p_1 \neq p_2 $

We use normal distribution because sample size is large enough.

$A_L = 0.95 + \cfrac{\alpha}{2} = 0.975  Z_{\mathit critical} = 1.96$

So, the **critical region** is (-1.96, 1.96)

$ Z_{\mathit{calculated}} = \cfrac{(\hat p_1 - \hat p_2) - (p_1 - p_2)}{\sqrt{\hat{p}\cdot(1-\hat{p})\cdot{(\frac{1}{n_1}+\frac{1}{n_2})}}} $

In which, 

$ \hat p = \cfrac{x_1+x_2}{n_1+n_2} $
$\hat p$ is called **pooled proportion**.

In our example, $\hat p = 0.04769$

$Z_{\mathit{calculated}} \approx -1.646 $

$Z_{\mathit{calculated}$ lies in the critical region. 

So, we **cant reject** the null hypothesis. 

With 5% significance level, the proportion of the two groups are the same.

###  Normal Distribution & Probability Examples

https://www.youtube.com/watch?v=gHBL5Zau3NE



$ P(a \leq x \leq b) = \int_{a}^{b} \cfrac{\cfrac{e^{-(x-\mu)^2}}{2\sigma^2}}{\sigma\sqrt{2\pi}} dx$

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x,n):
    u = x.mean()
    s = x.std()
    
    # divide [x.min(), x.max()] by n
    x = np.linspace(x.min(), x.max(), n)
    a = ((x - u) ** 2) / (2 * (s ** 2))
    y = 1 / (s * np.sqrt(2 * np.pi)) * np.exp(-a)
    
    return x, y, np.mean(y), np.std(y)

x = np.arange(-100, 100)
x, y, u, s = gaussian(x, 10000)

plt.plot(x, y, label=r'$\mu=%.2f,\ \sigma=%.2f$' % (u, s))
plt.legend()
plt.savefig('../figs/gaussian.png')
plt.show()
    
```
