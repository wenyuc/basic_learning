### Exponential distribution(continuous) 指数分布（连续型）
Exponential distribution is special cases of the gamma distribution when alpha is 1.

#### Characteristics
**Exponential Distribution** describes the time between events in a **Poisson process** (ie, the "inverse" of Poisson!)

For example, **Poisson** describes **the number of cars passing a toll gate in one hour (i.e., events per single unit of time).**
**Exponential** describes **the number of hours between car arrivals (i.e., time per single event).**

- It's continuous distribution
- The single parameter is $ \lambda = \cfrac{1}{\mu} $

#### P.D.F. (probability density function)
$$ f(x) = \lambda \cdot e^{-\lambda x} \;\;\; \lambda = \cfrac{1}{\mu} \;\;\; \mu = standard\; deviation\; for\; an\;  exponential\; distribution $$



$ E(X) = \cfrac{1}{\lambda} $

standard deviation $ s = \cfrac{1}{\lambda} $

#### C.D.F (cumulative density function)

$$ A_L = P(X \le x) = 1-e^{\left(\cfrac{-x}{\mu}\right)} = 1 - e^{-\lambda x} $$

$$ A_R = P(X > x) =  e^{-\lambda x} $$

#### Examples

1. Laptops produced by company XYZ last, on average, for 5 years. The life span of each laptop follows an exponential distribution.

**(a)** calculate the rate parameter.

$ \mu = 5\; years\; \lambda = \cfrac{1}{\mu} = \cfrac{1}{5} $

**(b)** what is the probability that a laptop will last less than 3 years?

$ A_L=P(X < x) =  1 - e^{-\lambda x} $

$ P(x<3) = 1-e^{-0.2\cdot 3} = 1-e^{-0.6} $

**(c)** what is the probability that a laptop will last more than 10 years?

$ A_R = P(X >x) = e^{-\lambda x} $

$ P(X>10) = e^{-0.20 \cdot 10} = e^{-2} $

https://www.youtube.com/watch?v=QKA4HNEw3aY

```python
import numpy as np
import matplotlib.pyplot as plt

def exponential(x, lamb):
    y = lamb * np.exp(-lamb * x)
    return x, y, np.mean(y), np.std(y)

for lamb in [0.5, 1, 1.5]:
    x = np.arange(0, 20, 0.01, dtype=np.float)
    x,y,u,s = exponential(x, lamb)
    plt.plot(x, y, label=r'$\mu=%.2f,\ \sigma=%.2f,'
                         r'\ \lambda=%d$' % (u, s, lamb))
plt.legend()
plt.savefig('../figs/exponential.png')
plt.show()
```
