### Uniform Distribution
$$ X \sim U(a,b) $$
**P.D.F.**
$$ f(x) = \cfrac{1}{b-a} \; a \leq x \leq b $$

**C.D.F.**

The probability is the area under the curve.

$$ P(X\leq x) = \cfrac{x-a}{b-a} $$



**mean and standard deviation**
$$ \mu = \cfrac{a+b}{2} \;\;\;\sigma = \cfrac{b-a}{\sqrt{12}} $$

#### Examples

https://www.youtube.com/watch?v=KfunVw-0AH0

1. The amount of time a person must wait for a train to arrive at a certin town is uniformly distributed between 0 and 40 minutes. 
**(a)** P(X=20) = **0**

**(b)** P(X > max) = **0**



```python
import numpy as np
import matplotlib.pyplot as plt

def uniform(x, a, b):
    
    y = [1 / (b - a) if (a<= val) and (val <= b)
        else 0 for val in x]
    
    return x, y, np.mean(y), np.std(y)

x = np.arange(-100, 100)    # define range of x
y = []

for zone in [(-50, 50), (10, 20)]:
    a, b = zone[0], zone[1]
    #print('a=%d b = %d', a, b)
    x, y, u, s = uniform(x, a, b)
    plt.plot(x,y, label= r'$\mu=%.2f,\ \sigma = %.2f$' % (u, s))

plt.legend()
plt.savefig('../figs/uniform.png')
plt.show()
```
