逻辑斯谛回归(logistic regression)是统计学习中的经典分类方法。

最大熵(maximum entropy)是概率模型学习的一个准则，将其推广到分类问题得到最大熵模型。

逻辑斯谛回归与最大熵模型都属于**对数线性模型**。

### 逻辑斯谛回归模型

#### 逻辑斯谛分布(logistic distribution)

定义6.1(Logistic Distribution) 设$ \mathit X$是连续随机变量，X服从逻辑斯谛分布是指$ \mathit X$具有下列分布函数和密度函数：

$$ \mathit F(x) = P(X \leq x) = \cfrac{1}{1 + e^{-\frac{(x-\mu)}{\gamma}}} \;\;\; (6.1)$$

$$ \mathit f(x) = F\prime(x) = \cfrac{e^{-\frac{(x-\mu)}{\gamma}}}{\gamma(1 + e^{-\frac{(x-\mu)}{\gamma}})} \;\;\; (6.2)$$

其中，$\mu$为位置参数，$\gamma$为形状参数。



逻辑斯谛分布的密度函数$\mathit f(x)$和分布函数$ \mathit F(x) $的图形如上所示。**分布函数属于逻辑斯谛函数，其图形是一条S形曲线(sigmoid curve)**。

该曲线以点$\mathit (\mu, \cfrac{1}{2})$为中心对称，即满足

$$ \mathit F(-x + \mu) - \cfrac{1}{2} = - F(x+\mu) + \cfrac{1}{2} $$

曲线在中心附近增长速度较快，在两端增长速度较慢。
形状参数$\gamma$ 的值越小，曲线在中心附近增长得越快。



#### 二项逻辑斯谛回归模型

二项逻辑斯谛回归模型(ninomial logistic regression model)是一种分类模型，由条件概率$ \mathit P(Y \mid X) $表示，形式为参数化的逻辑斯谛分布。这里，随机变量$ \mathit X$ 取值为实数，随机变量$ \mathit Y$取值为1或0。我们通过监督学习方法来估计模型参数。

定义6.2（**逻辑斯谛回归模型 Logistic Regression Model**) 二项逻辑斯谛回归模型是如下的条件概率分布：

$$ \mathit P(Y=1\mid x) = \cfrac{e^{\omega\cdot x + b}}{1 + e^{\omega\cdot x + b}} \;\;\; (6.3) $$

$$ \mathit P(Y=0\mid x) = \cfrac{e^{\omega\cdot x + b}}{1 + e^{\omega\cdot x + b}} \;\;\; (6.4) $$

这里，$\mathit x\in R^n $是输入， $\mathit Y \in \{0,1\}$是输出。$ \mathit \omega \in R^n $ 和 $ \mathit b \in R$是参数，$\omega$称为权值向量，$ \mathit b$ 为偏置，$\mathit \omega \cdot x$是$\omega$和$\mathit b$的**内积**。

逻辑回归比较两个条件概率值的大小，将实例$\mathit x$分到概率值较大的类。

---

有时为了方便，将权值向量和输入向量加以扩充，仍记作$\mathit \omega \; , x$,即
$$ \mathit \omega =(\omega^{(1)}, \omega^{(2)}, \cdots, \omega^{(n)}, b)^T , x = (x^{(1)}, x^{(2)}, \cdots, x^{(n)}, 1)^T  $$

这时，逻辑斯谛回归模型如下：
$$ \mathbf P(Y=1\mid x) = \cfrac{e^{\omega\cdot x}}{1 + e^{\omega\cdot x}} \;\;\; (6.5) $$

$$ \mathbf P(Y=0\mid x) = \cfrac{e^{\omega\cdot x}}{1 + e^{\omega\cdot x}} \;\;\; (6.6) $$



#### 考察逻辑斯谛回归模型的特点。

一个事件的**几率(odds)** 是指该事件发生的**概率(probability)** 与该事件不发生的**概率**的比值。

如果事件发生的概率是$\mathit p$，那么该事件的**几率** 是$ \cfrac{p}{1-p}$,该事件的**对数几率** (log odds)或 logit函数是：

$$ \mathit logit(p) = log \frac{p}{1-p} $$

对逻辑斯谛回归而言，由式(6.5)(6.6)得

$$ log \cfrac{P(Y=1\mid x)}{1 - P(Y=1\mid x)} = \omega \cdot x $$

就是说，在逻辑斯谛回归模型中，输出Y = 1的对数几率是输入 x 的线性函数。

或者说，输出Y = 1的对数几率是由输入 x 的线性函数表示的模型，即逻辑斯谛回归模型。

通过逻辑斯谛回归模型的定义式(6.5)可以将线性函数$\omega \cdot x$ 转换为概率：

$$ \mathbf P(Y=1\mid x) = \cfrac{exp(\omega\cdot x)}{1 + exp(\omega\cdot x)} $$

这时，线性函数的值越接近正无穷,概率值就越接近1；线性函数的值越接近负无穷,概率值就越接近0.

这样的模型就是逻辑斯谛回归模型。



#### 逻辑斯谛回归模型参数估计

逻辑斯谛回归模型学习时，对于给定的训练数据集 $ \mathit T = \{(x_1, y_1), (x_2, y_2), \cdots, (x_N, y_N)\}$,
其中，$ x_i \in R^n, y_i \in \{0,1\} $, 可以用**极大似然估计法估计模型参数，从而得到逻辑斯蒂回归模型**。

推到见书 P93

逻辑斯谛回归学习中通常采用的方法时梯度下降法及拟牛顿法。

假设$\omega$的极大似然估计值是$\hat \omega$，那么学到的逻辑斯蒂回归模型为

$$ \mathbf P(Y=1\mid x) = \cfrac{exp(\hat\omega\cdot x)}{1 + exp(\hat\omega\cdot x)} $$

$$ \mathbf P(Y=0\mid x) = \cfrac{1}{1 + exp(\hat\omega\cdot x)} $$

#### 多项逻辑斯蒂回归

可以将其推广为多项逻辑斯蒂回归模型(multi-nominal logistic regression model)，用于多类分类。

假设离散型随机变量$\mathit Y$的取值集合是$\{1,2,\cdots, K\}$,那么**多项逻辑斯蒂回归模型**是

$$ \mathbf P(Y=k\mid x) = \cfrac{exp(\omega_k\cdot x)}{1 + \sum_{k=1}^{K-1}exp(\omega_k\cdot x)} \;\; k=1,2,\cdots, K-1 \;(6.7) $$

$$ \mathbf P(Y=K\mid x) = \cfrac{1}{1 + \sum_{k=1}^{K-1}exp(\omega_k\cdot x)} \;\; k=1,2,\cdots, K-1 \;(6.8) $$

这里，$x \in R^{n+1}, \omega_k \in R^{n+1} $。

二项逻辑斯蒂回归的参数估计发也可以推广到多项逻辑斯谛回归。


### 最大熵模型

#### 最大熵原理

最大熵原理是概率模型学习的一个准则。

最大熵原理认为，**在学习概率模型时，在所有可能的概率模型（分布）中，熵最大的模型是最好的模型**。

通常用约束条件来确定概率模型的集合，所以，最大熵原理也可以表述为在**满足约束条件的模型集合中选取熵最大的模型**。


假设离散随机变量$\mathit X$的概率分布是$\mathit P(x)$，则其熵是

$$ \mathit H(P) = - \sum_{x} P(x)log P(x)   \;\;\; (6.9) $$

熵满足下列不等式:

$$ 0 \leq H(P) \leq log \mid X \mid $$

式中，$\mid X \mid$是X的取值个数。当X服从均匀分布时，熵最大。

#### 最大熵模型定义

最大熵原理是统计学习的一般原理，将它应用到分类得到最大熵模型。

假设分类模型是一个条件概率分布$\mathit P(Y\mid X), X \in \cal{X} \subseteq R^n $表示输入，$Y \in \cal Y$ 表示输出。

这个模型表示的是对于给定的输入$\mathit X$，以条件概率$\mathit P(Y \mid X)$输出Y。

给定一个训练数据集

$$ \mathit T = \{(x_1, y_1), (x_2, y_2), \cdots, (x_N, y_N)\}$$

学习的目的是用最大熵原理选择最好的分类模型。

首先考虑模型应该满足的条件。给定训练数据集，可以确定联合分布$\mathit P(X,Y)$的经验分布和边缘分布$\mathit P(X)$的经验分布，分别以$\tilde P(X,Y), \tilde P(X) $表示，这里，

$$ \tilde P(X=x, Y=y) = \cfrac{\nu(X=x, Y=y)}{N} $$

$$ \tilde P(X=x) =  \cfrac{\nu(X=x)}{N} $$

其中，$\nu(X=x, Y=y)$表示训练数据中样本(x,y)出现的频数， 其中，$\nu(X=x)$表示训练数据中输入x出现的频数，N表示训练样本容量。

用**特征函数(feature function)** $\mathit f(x,y)$描述输入x和输出y之间的某一个事实。其定义是：

$$ \mathit
f(x,y) = 
\begin{cases}
1,&x与y满足某一事实 \\
0,&否则 
\end{cases}
$$

它是一个二值函数。

特征函数f(x,y) 关于 经验分布 $ \tilde P(X=x, Y=y)$的期望值，用$E_{\tilde P}(f)$表示

$$ E_{\tilde P}(f) = \sum_{x,y} \tilde P(x, y)f(x,y) $$

特征函数f(x,y) 关于模型 $ P(Y \mid X)$与经验分布$\tilde P(x)$的期望值，用$E_{P}(f)$表示

$$ E_{P}(f) = \sum_{x,y} \tilde P(x)P(y|x)f(x,y) $$

**定义 6.3（最大熵模型）** 假设满足所有约束条件的模型集合为

$$ \cal C \equiv \mathit \{P \in \cal P \mid\mathit E_{P}(f_i) = \mathit E_{\tilde P}(f_i)\;  i=1,2,\cdots, n\} （6.12）$$

定义在条件概率分布$\mathit P(Y|X)$上的条件熵为

$$ H(P) =  -\sum_{x,y} \tilde P(x)P(y|x)logP(y|x)\;\;\; (6.13) $$

则模型集合$\cal C$中条件熵H(P)最大的模型称为最大熵模型。式中的对数为自然对数。
