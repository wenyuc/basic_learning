### Logistic Regression（逻辑回归） Introduction
 逻辑回归是在线性回归的基础上加了一个Sigmoid函数（非线性）映射，使得逻辑回归称为了一个优秀的分类算法。
 
 逻辑回归假设数据服从伯努利分布，通过极大化似然函数的方法，运用梯度下降来求解参数，来达到将数据二分类的目的。
 
 相比于协同过滤和矩阵分解利用物品“相似度”进行推荐，逻辑回归将问题看成一个分类问题，通过预测正样本的概率对物品进行排序。
 因此，逻辑回归模型将推荐问题转化成了一个点击率预估问题。
 
 #### LR 的优点
 1. LR模型形式简单，可解释性好，从特征的权重可以看到不同的特征对最后结果的影响。
 2. 训练时便于并行化，在预测时只需要对特征进行线性加权，所以性能比较好，往往适合处理**海量id类特征**，用id类特征
 的好处就是**防止信息损失**。
 3. 资源占用小，尤其是内存。
 4. 方便输出结果调整。逻辑回归可以很方便的得到最后的分类的结果，因为输出的是每个样本的概率分数，可以很容易的对这些概率
 分数进行cutoff，也就是划分阈值。
 
 #### LR的局限性
 1. 表达能力不强，无法进行特征交叉和特征筛选灯一系列“高级”操作；
 2. 准确率并不是很高，毕竟是一个线性模型加一个Sigmoid，形式简单，很难去拟和数据的真实分布；
 3. 处理非线性数据麻烦，LR在不引入其他方法下，只能处理线性可分数据。
 4. LR需要进行人工特征组合。
 
 这些缺点可通过GBDT模型化解

### Logistic Regression
https://www.youtube.com/watch?v=zAULhNrnuL4

**Logistic regression seeks to:**
- **model** the probability of an event occurring depending on the values of
the independent variables, which can be categorical or numerical;
- **estimate** the probability that an event occurs for a randomly selected
observation versus the probability that the event does not occur. 
- **predict** the effect of a series of variables on a binary response variable.
- **classify** observations by estimating the probability that an observation is
in a particular category (such as approved or not approved in our problem.)

**What are probability**
P = (outcomes of interest) / (all possible outcomes)

**what are odds**

odds = P(occuring) / P(not occurring)

odds = p / (1-p)

Fair coin flip  odds(heads) = 0.5 / (1-0.5) = 1
fair die roll   p(1 or 2) = 2/6 = 1/3
                odds(1 or 2) = (1/3) / (1-1/3) = 1/2
deck of playing cards odds (diamond card) = 0.25/0.75 = 1/3

**What is odds ratio**
The odds ratio is exactly a ratio of two odds.
*Fair coin flip*
p(heads) =0.5
odds(heads) = 0.5 / 0.5 = 1

*loaded coin flip*
p(heads) = 0.7
odds(heads) = 0.7/0.3 = 2.3333

odds ratio = odds1 / odds0 = 2.33333

This means that:
*The odds of getting "heads" on the loaded coin are
2.333x greater than the fair coin.*

**odds ratio in logistic regression**
- The odds ratio for a variable in logistic regression represents
how the odds change with a 1 unit increase in that variable 
holding all other variables constant. 

- For example:
    - Body weight and sleep apnea (two categories: apnea / no apnea)
    - weight variable had an odds ratio of 1.07
    - This means a one pound increase in weight increases the odds of 
      having sleep apnea by 1.07 (not very high b/c we are looking at 
      1lb increment)
    - A ten pound increase in weight increases the odds of 1.98, or almost
      doulbes a person's odds of having sleep apnea and a 20 pound increase
      raises the odds to 3.87.
    - This holds true at any point in the weight spectrum. From 200 pound to
      201 pound, the odd ratio is 1.07, while, from 150 to 151 pound, 
      the odd ratio is still 1.07
 
 **A warning**
 - It is very important **to separate probability and odds**
 - In the previous example, a person gaining 20 pounds increases,
   their odds of sleep apnea by almost a factor 4 
   **regardless of their starting weight**
 
 - However, the **probability of having sleep apnea is LOWER**
   in person with lower body weight to begin with

- So while the odds are 4x greater, the probability may still be low
- Basically what this means is that the odds can have a large 
  magnitude even if the underlying probabilities are low.
  
#### Bring Back Bernoulli

- The dependent variable in logistic regression follows the Bernoulli
distribution having an unknown probability, p

- Remember that the Bernoulli distribution is just a special case of
the Binomial distribution where n=1 (just one trial)

- success is "1" and failure is "0"
- So the probability of success is *p* and failure is *q=1-p*
- **In logistic regression, we are estimating an unknown p for any given linear combination of the independent variables.**
- Therefore we need to link together our independent variables to essentially
the Bernoulli distribution, that link is called the **logit**

#### What is the LOGIT
https://www.youtube.com/watch?v=NmjT1_nClzg
- In logistic regression, we do not know *p* like we do in Binomial (Bernoulli) distribution problems. The goal of logistic regression is to estimate *p* for a linear combination of the independent variables. Estimate of p is p-hat. 

- To tie together our linear combination of variables and in essence the Bernoulli distribution we need a function that links them together, or maps the linear combination of variables that could result in any value onto the Bernoulli probability distribution with a domain from 0 to 1. The natural log of the odds ratio, **the logit**, is that link function. 
ln(odds) => ln(p/(1-p)) is the logit(p), or ln(p) - ln(1-p) = logit(p)
loge(x) = ln(x)


 
ln(0.5/0.5) = ln(1) = 0
we call an s-curve, or a sigmoid curve. 

**What is the inverse logit**
In our logit link function graph, 0 to 1 ran along the x-axis but we want the probabilities to be on the y-axis. We can achieve that by taking the **inverse logit function**.

logit(p) = ln(p/(1-p)), where p is between 0 and 1

logit-1(a) = 1/(1+e^(-a)) = e^a / (e^a+1), where a is some number

In our case, some *number* will be the linear combination of variables and their coefficients. The inverse-logit will return the probability of being a "1" or in the "event occurs" group.


Mean function
u y|x = e^a / (e^a+1)

The graph of inverse logit


**Sigmoid function converts input into range 0 to 1.**

**About Coefficients**
- The regression coefficients for logistic regression are calculated using **maximum likelihood estimation or MLE. **
- MLE does tie a few things together behind the scenes, so for now we will jsut let id do its thing. 

**Estimated Regression Equation**
- The natural logarithm of the odds ratio is equivalent to a **linear** function of the independent variables. Tha antilog of the logit function allows us to find the estimated regression equation.

logit(p) = ln(p/(1-p)) = b0 +b1x1
antilog =>
p/(1-p) = e^(b0+b1x1)
=>
p-hat = e^(b0+b1x1) / (1+ e^(b0+b1x1))
This is the estimated regression equation.




### How to perform Logistic Regression in Python
https://www.youtube.com/watch?v=HYcXgN9HaTM

```python
# step 1: Import the required modules
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd 
```

```python
# step 2: Generate the dataset
# generate a dataset for LogisticRegression
x,y = make_classification(n_samples=100, n_features=1, n_classes=2,\
                         n_clusters_per_class=1, flip_y=2, \
                         n_informative=1, n_redundant=0, \
                         n_repeated=0)
```

```python
x.shape
```

```python
y
```

```python
# step 3 visualize the data
# create a scatter plot
plt.scatter(x,y, c=y, cmap='rainbow')
#plt.scatter(df.age, df.bought_insurance, marker='+', color='red')

plt.title("Scatter Plot of Logistic Regression")
plt.show()
```

```python
# Step 4: Split the dataset
# Split the dataset into training and test dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 0)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# step 5: Perform Logistic Regression
log_reg=LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
# show the coefficient and Intercept
print(log_reg.coef_)
print(log_reg.intercept_)
```

```python
# step 6: Make prediction using the model

y_pred = log_reg.predict(x_test)
```

```python
# step 7: Display the Confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
#step 7.1 calculate the score
log_reg.score(x_test, y_test)
```

```python
# step 7.2 calculate the probability, no_prob, yes_prob
log_reg.predict_proba(x_test)
```
