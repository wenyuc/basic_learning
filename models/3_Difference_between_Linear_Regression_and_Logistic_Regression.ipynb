### Difference Between Linear Regression and Logistic Regression

- Outcome

In linear regression, the outcome (dependent variable) is continuous. It can have any one of an infinite number of possible values.

In logistic regression, the outcome (dependent variable) has only a limited number of possible values. 

- The dependent variable

Logistic regression is used when the response variable is categorical in nature. For instance, yes/no, true/false, red/green/blue, 1st/2nd/3rd/4th, etc.

Linear regression is used when your response variable is continuous. For instance, weight, height, number of hours, etc.

- Equation

Linear regression gives an equation which is of the form $Y = b_0+b_1x + c$

However, logistic regression gives an equation which is of the form $ log\left(\cfrac{Y}{1-Y}\right) = C + b_1x_1+b_2x_2+\cdots $

- Estimation Method

Linear regression uses **least square estimation**.

Logistic regression uses **maximum like-hood estimation**.

- Best fit line

**Straight line** for linear regression.

**Curve line** for logistic regression.

- Relationship between DV and IV

Linear relationship between dependent variable (dv) and independent variable (iv) for linear regression.

Linear relationship is not mandatory for logistic regression.
