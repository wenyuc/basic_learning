### 画一个标准的坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create figure window
plt.figure(num=3, figsize=(8, 5), facecolor='white', edgecolor='yellow')

# plot y1 and y2
plt.plot(x, y1)
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='--')

# set axes range
plt.xlim((-5, 5))
plt.ylim((-2, 2))

# set axes label
plt.xlabel("Data Sets", fontsize=13, fontweight='bold')
plt.ylabel("Accuracy", fontsize=13, fontweight='bold')

# set axes ticks 坐标轴刻度
my_x_ticks = np.arange(-5, 5, 0.5)
my_y_ticks = np.arange(-2, 2, 0.3)
plt.xticks(my_x_ticks, fontsize=8, color='red', rotation=45)
plt.yticks(my_y_ticks, fontsize=8, color='red')

# don't show axes ticks 不显示坐标刻度
# plt.xticks([])
# plt.yticks([])
# ax.set_xticks([])
# ax.set_yticks([])

plt.axis('tight')
plt.show()
```

### x-轴刻度上显示为labels
#### 参数x, labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(1,10, 1)
y = range(1,10, 1)

# plt.xticks()中第一个参数locs 表示x-axis刻度线显示标注的地方，即ticks放置的地方
# 第二个参数表示在locs数组表示的位置添加的标签，labels不赋值时，标签即为locs，如赋空值()则无ticks
plt.xticks(x, labels=['a','b','c','d','e','f','g','h','i'])
plt.plot(x, y)
plt.show()
```

#### 除了x, labels之外，常用的color，rotation

```python
import matplotlib.pyplot as plt
import numpy as np
import calendar

x = range(1,13,1)
y = range(1,13,1)

# ticks上的标签可以赋予任意其他的值
plt.xticks(x, calendar.month_name[1:13], color = 'blue',rotation = 60)
plt.plot(x,y)
plt.show()
```

### y轴的log点

```python
import matplotlib.pyplot as plt
import numpy as np

# create data between (0,1)
y = np.random.normal(loc=0.5, scale= 0.4, size = 1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))

# plot
# plt.yscale('linear')
plt.yscale('log')
# plt.yscale('symlog', linthreshy = 0.05)
# plt.yscale('logit')

plt.show()
```

### x轴的log点

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

x = np.arange(0.0, 100.0, 1)
y = np.sin(0.1*np.pi*x)*np.exp(-x * 0.01)

ax = plt.subplot(111)

plt.plot(x, y, '--r*')
plt.semilogx()  # 半对数坐标系
print(plt.Axes.get_xscale)
plt.show();
```
