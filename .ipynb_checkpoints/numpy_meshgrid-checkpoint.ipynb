```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
%matplotlib inline
```

#### Ex 1

```python
x = np.array([0,0.5,1])
y = np.array([0,1,2])
xv,yv = np.meshgrid(x,y)

print('xv:{},xv的维度:{},shape{}'.format(xv, xv.ndim,xv.shape))
print('yv:{},yv的维度:{},shape{}'.format(yv,yv.ndim,yv.shape))
plt.plot(xv,yv, 'o--')
plt.grid(True)
plt.show();
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

x = np.arange(-10,11,1)
y = np.arange(-10,11,1)
xx, yy = np.meshgrid(x, y) #二维矩阵
fig = plt.figure(1, figsize=(12,8))
zz = (xx**2 + yy**2)
ax = fig.add_subplot(1,1,1,projection='3d')
ax.set_top_view()

ax.plot_surface(xx,yy,zz,rstride=1,cstride=1,cmap='rainbow')
plt.show();
```
