```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. simple plot with 4 numbers

```python
plt.plot([1,3,2,4])
plt.show()
```

# 2. points have x and y values, add title and axis labels

```python
plt.plot([1,2,3,4], [1,4,9, 16])
plt.title("Test Plot", fontsize = 8, color = 'g')
plt.xlabel('number n')
plt.ylabel('n^2')
plt.show()
```

# 3. change figure size, plot red dots, set axis scales x: 0-6 and y: 0-20

```python
plt.figure(figsize = (2,5))
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.annotate('square it', (3,6))
plt.show()
```

# 4. Bar chart with four bars


```python
plt.clf()    # clear figure
x = np.arange(4)
y = x ** 2 + 1
plt.xticks(x, ('Ankit', 'Hans', 'Joe', 'Flanc'))
plt.bar(x, y, color = ['lime', 'r', 'k', 'tan'])
plt.show()
```

# 5. Two sets of 10 random dots plotted

```python
d = {'Red O': np.random.rand(10),
    'Grn X': np.random.rand(10)}
df = pd.DataFrame(d)
df.plot(style = ['ro', 'gx'])
plt.legend()
plt.show()
```

# 6. Time series - six months of random floats

```python
ts = pd.Series(np.random.randn(182), 
              index = pd.date_range('1/1/2020', periods = 182))
df = pd.DataFrame(np.random.randn(182, 3), index= ts.index,
                  columns = list('ABC'))
df.cumsum().plot()
plt.show()
```

# 7. Draw Quadratics

```python
plt.style.available
```

```python
plt.style.use('seaborn-deep')
x = np.linspace(-20, 20, 1000)
y = [2 * i**2 -5* i + 4 for i in x]
plt.plot(x,y)
plt.show()
```

# Draw polynominals

```python
x = np.linspace(-30, 30, 2000)
y1 = [i**6 for i in x]
y2 = [i**5 for i in x]
y3 = [i**4 for i in x]
y4 = [i**3 for i in x]
y5 = [i**2 for i in x]
y6 = [i for i in x]

plt.plot(x, y1, color = 'r', label = 'x^6')
plt.plot(x, y2, color = 'g', label = 'x^5')
plt.plot(x, y3, color = 'b', label = 'x^4')
plt.plot(x, y4, color = 'black', label = 'x^3')
plt.plot(x, y5, color = 'k', label = 'x^2')
plt.plot(x, y6, color = 'tan', label = 'x')

plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(-30, 30, 2000)
y1 = [i for i in x]

ax = plt.axes()
ax.plot(x, y1, '-r')
ax.set(xlim = (-35, 35), ylim = (-40, 40), xlabel = 'X', ylabel ='Y',
      title ='x')
```

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(-30, 30, 2000)
y2 = [i**2 for i in x]

ax = plt.axes()
ax.plot(x, y2, '-g', label = 'x^2')
ax.set(xlim = (-35, 35), ylim = (-10, 1000), xlabel = 'X', ylabel ='Y',
      title ='x^2')

plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(-30, 30, 2000)
y3 = [i**3 for i in x]

ax = plt.axes()
ax.plot(x, y3, '-g')
ax.set(xlim = (-35, 35), ylim = (-10000, 10000), xlabel = 'X', ylabel ='Y',
      title ='x^3')
```

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(-30, 30, 2000)
y4 = [i**4 for i in x]

ax = plt.axes()
ax.plot(x, y4, '-b', label = 'x^4')
ax.set(xlim = (-35, 35), ylim = (-10, 100000), xlabel = 'X', ylabel ='Y',
      title ='x^4')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.style.use('seaborn-whitegrid')
x = np.linspace(-30, 30, 2000)
y5 = [i**5 for i in x]

ax = plt.axes()
ax.plot(x, y5, '-b', label = 'x^4')
ax.set(xlim = (-35, 35), ylim = (-100000, 100000), xlabel = 'X', ylabel ='Y',
      title ='x^5')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.style.use('seaborn-whitegrid')
x = np.linspace(-20, 20, 2000)
y = [5*i**4 - 4*i**3 +10 for i in x]

ax = plt.axes()
ax.plot(x, y, '-r', label = '5*x^4 - 4*x^3 +10')
ax.set(xlim = (-25, 25), ylim = (-100, 1000), xlabel = 'X', ylabel ='Y',
      title ='5*x^4 - 4*x^3 +10')

plt.legend()
plt.show()
```
