#### Making Change to Data

```python
poke.head(5)
```

```python
# Add a new column
poke["Total"] = poke["HP"] + poke["Attack"] + poke["Defense"] + poke["Sp. Atk"] + poke["Sp. Def"] + poke["Speed"]
poke.head()
```

```python
poke = poke.drop(columns = ['Total'])
poke.head()
```

```python
poke["Total"] = poke.iloc[:, 4:10].sum(axis = 1)
poke.head(5)
```

```python
45+49+49+65+65+45
```

```python
poke2 = poke[['Total', "HP", "Defense"]]
poke2.head()
```

```python
cols = list(poke.columns.values)
poke2 = poke[cols[0:4] + [cols[-1]] + cols[4:12]]
poke2.head()
```

#### Saving as csv

```python
poke2.to_csv("./data/pokemon_modified.csv", index = False)
```

#### Filtering Data

```python
poke.shape
```

```python
poke2 = poke.loc[(poke["Type 1"] == "Grass") & (poke["Type 2"] == "Poison") & (poke["HP"] > 70)]
poke2
```

```python
# by default, it saves the old index column. use drop parameter to drop the old index column
poke2 = poke2.reset_index(drop = True)
poke2  
```

```python
poke.loc[~poke["Name"].str.contains('Mega')]
```

```python
import re

poke.loc[poke["Type 1"].str.contains('fire|grass', flags = re.I, regex = True)]
```

```python
poke.loc[poke["Name"].str.contains('pi[a-z]*', flags = re.I, regex = True)]
```

```python
poke.loc[poke["Name"].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

#### Conditional Changes

```python
#poke.loc[poke["Type 1"] == "Fire", "Type 1"] = "Flame"
poke.loc[poke["Type 1"] == "Fire", "Legendary"] = True
poke
```

```python
poke.loc[poke["Total"] > 500, ['Generation', 'Legendary']] = ["Test 1", "Test 2"]
poke
```

#### Aggregate Statistics (Groupby)

```python
poke.groupby(["Type 1"]).mean().sort_values('HP', ascending = False)
```

```python
poke.groupby("Type 1").count()
```

```python
poke['count'] = 1
poke
```

```python
poke.groupby(["Type 1", "Type 2"]).count()['count']
```

#### Working with a large amount of data

```python
for df in pd.read_csv("./data/pokemon_modified.csv", chunksize = 5):
    print(df)
```

```python
new_pokemon = pd.DataFrame(columns = poke.columns)
for df in pd.read_csv("./data/pokemon_modified.csv", chunksize = 5):
    results = df.groupby(["Type 1"]).count()
    new_pokemon = pd.concat([new_pokemon, results])

new_pokemon
    
```
